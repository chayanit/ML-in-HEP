# uproot

```python
import uproot
import awkward
import numpy as np
import pandas as pd
```

```python
signal = uproot.open('SUSYGluGluToBBH_M-300_Delphes.root')
signal
```

```python
signal_t = signal['mssmhbb']
signal_t
```

```python
signal_b = signal_t.arrays()
```

```python
signal_df = awkward.to_pandas(signal_b)
```

```python
signal_df
```

```python
np.ones(len(signal_df))
```

```python
signal_df.loc[:,'label'] = np.ones(len(signal_df)).tolist()
```

```python
signal_df
```

```python
signal_df.describe()
```

```python
bg = uproot.open('QCD_bEnriched_HT100toInf_Delphes.root')
bg
```

```python
bg_t = bg['mssmhbb']
bg_t
```

```python
bg_b = bg_t.arrays()
```

```python
bg_df = awkward.to_pandas(bg_b)
```

```python
bg_df
```

```python
bg_df.describe()
```

```python
bg_df.loc[:,'label'] = np.zeros(len(bg_df)).tolist()
```

```python
bg_df
```

```python
dataset_df = pd.concat([signal_df,bg_df], ignore_index=True)
```

```python
dataset_df
```

```python
dataset_df.to_csv('dataset_Delphes.csv')
```
